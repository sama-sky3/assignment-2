```python
import pandas as pd
```

```python
data = pd.read_csv('archive/books.csv')
```

```python
sel_cols = ['title', 'authors', 'original_publication_year', 'average_rating', 'ratings_count']
df = data[sel_cols]

# get Harry Potter books written by J.K. Rowling
# Get unique authors from the 'authors' column
authors = df.authors.unique()

# Filter authors to include only those written by J.K. Rowling
jkr = [i for i in authors if i.find('J.K. Rowling') != -1]

# Filter the DataFrame to include only books written by J.K. Rowling
jkr_books = df.loc[df.authors.isin(jkr)]

# Get unique titles from the filtered DataFrame
all_titles = jkr_books.title.unique()

# Filter titles to include only those containing 'Harry Potter and '
hp_titles = [i for i in all_titles if i.find('Harry Potter and ') != -1]

# Filter the DataFrame to include only Harry Potter books written by J.K. Rowling
hp_books = jkr_books.loc[jkr_books.title.isin(hp_titles)]

# Sort the DataFrame by the original publication year in ascending order
hp_books = hp_books.sort_values(by=['original_publication_year'], ascending=True)

display(hp_books)
```
